#### Notebook to explore benchmark data

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('datasets/A1Benchmark/real_1.csv', index_col='timestamp')
df_anomalies = df[df['is_anomaly']==1]

# Create figure and plot space
fig, ax = plt.subplots(figsize=(12, 5))

# Add x-axis and y-axis
ax.plot(df.index,
        df['value'],
        color='blue',
       label = 'All Values')

ax.plot(df_anomalies.index,
        df_anomalies['value'],
        'ro',
       label='Labeled Anomalies')


# Set title and labels for axes
ax.set(xlabel="Timestamp",
       ylabel="Value",
       title="Yahoo Dataset- A1 Real 1")

plt.legend(loc="upper left")

plt.show()
```
